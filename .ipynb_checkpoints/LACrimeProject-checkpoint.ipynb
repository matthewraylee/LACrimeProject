# Exploring LA Crimes

## Import necessary Libraries and Loading Data

```python
import pandas as pd
import matplotlib.pyplot as plt
la_crime = pd.read_csv('./Crime_Data_from_2020_to_Present_20240923.csv')
```

## Understanding and Cleaning Data

```python
la_crime.head()
```

```python
la_crime.columns
```

```python
# Check null values
la_crime.isna().sum()
```

### Drop all columns that contain null values

```python
df = la_crime.drop(labels = [
    'Mocodes', 
    'Vict Sex', 
    'Vict Descent', 
    'Weapon Used Cd', 
    'Weapon Desc', 
    'Crm Cd 2', 
    'Crm Cd 3', 
    'Crm Cd 4', 
    'Cross Street'
], axis=1)
```

```python
df.head()
```

### Sort crimes from earliest date to present

```python
df.sort_values(by='Date Rptd')
```

### Update data types

```python
df.dtypes
```

```python
df.head()
```

```python
df["Date Rptd"] = pd.to_datetime(df["Date Rptd"], format='%m/%d/%Y %I:%M:%S %p')
df["DATE OCC"] = pd.to_datetime(df["DATE OCC"], format='%m/%d/%Y %I:%M:%S %p')
df["TIME OCC"] = pd.to_datetime(df["TIME OCC"].astype(str).str.zfill(4), format='%H%M').dt.time
```

```python
df.info()
```

### Summary of table

```python
# Print description of numbers in table
df_desc = df.describe(include = ['int64', 'float64'])
df_desc
```

```python
df['Vict Age'].value_counts()
```

There are negative values in age column

```python
# Remove negative values for age
df_age = df[df['Vict Age'] > 0]
df_age['Vict Age'].value_counts()
```

### Add Year and Month Column

```python
# Make new column year
df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month
df
```

## Feature Understanding

```python
df['Month'].value_counts().sort_index().plot()
```

```python
df['Year'].value_counts().sort_index().plot(kind='bar')
```

```python
# Remove incomplete month: Sept 2024
df
df = df[~((df['Month'] == 9) & (df['Year'] == 2024))]
df.sort_values(by='DATE OCC')
```

```python
crime_overtime = df.groupby(['Year', 'Month']).size().reset_index(name='Count')
crime_overtime['Year-Month'] = pd.to_datetime(crime_overtime['Year'].astype(str) + '-' + 
                                                crime_overtime['Month'].astype(str))
crime_overtime.head()
```

```python
plt.plot(crime_overtime['Year-Month'], crime_overtime['Count'])
plt.title("LA Crime Rate Over the Years")
```

### Explore age

```python
plt.hist(df_age['Vict Age'], edgecolor='black', bins=15)
plt.title('Number of Victim by Age Group')
plt.show()
```

### Explore Crime Count and Type per Day

```python
crime_count_per_day = df.groupby(['DATE OCC', 'Crm Cd Desc']).size()
crime_count_per_day
```

```python
top_20_crimes = crime_count_per_day.groupby('Crm Cd Desc').mean()\
.reset_index(name = 'Avg Daily Count').sort_values(by='Avg Daily Count', ascending=False).head(20)
top_20_crimes
```

```python
top_20_crimes_sorted = top_20_crimes.sort_values(by='Avg Daily Count', ascending=True)
```

```python
plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['Avg Daily Count'])
plt.title('Top 20 Crimes in LA')
plt.show()
```

### Summary of Plots

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,8))

# Line Chart
plt.subplot(2, 2, 1)
plt.plot(crime_overtime['Year-Month'], crime_overtime['Count'], c='green', linewidth='2')
plt.xlabel('Year-Month')
plt.title("LA Crime Over Time")

# Histogram
plt.subplot(2, 2, 2)
plt.hist(df_age['Vict Age'], edgecolor='black', bins=15)
plt.xlabel('Age')
plt.title('Distribution of Victim Age')

# Bar Chart
plt.figure()
plt.barh(top_20_crimes_sorted['Crm Cd Desc'], top_20_crimes_sorted['Avg Daily Count'], color='orange')
plt.ylabel('Type of Crime')
plt.title('Top 20 Crimes in LA')
plt.show()
```

```python
# Check the values of the column Time to Report
df.head()
```

```python
df['Time to Report'].value_counts().head(10)
```

```python
# Show the average amount of time to report based on the crime 
df.groupby(['Crm Cd Desc'])['Time to Report'].mean().sort_values()
```

## Areas with Most Crime During Night Time

```python
df["TIME OCC"] = pd.to_datetime(df["TIME OCC"], format='%H:%M:%S')
```

```python
df.head()
```

```python
df_night_time = df[(df['TIME OCC'].dt.hour >= 20) | (df['TIME OCC'].dt.hour <= 3)]
df_night_time
```

```python
df_night_time.groupby(['AREA NAME']).size().sort_values(ascending=False)
```

The above shows that Central has the highest Crime Rate at night

## Crime Count by Time of Day

```python
hourly_crime = df['TIME OCC'].dt.hour.value_counts().sort_index().reset_index()
hourly_crime
```

```python
plt.bar(hourly_crime['TIME OCC'], hourly_crime['count'])
plt.title('Crime Count by Time of Day')
plt.show()
```

There seems to be an anomaly, and we can check to make sure what's going on over there

```python
df[df['TIME OCC'].dt.hour == 12]['Crm Cd Desc'].value_counts()
```

Theft of Identity may be inputted at 12 as default. We will remove this.

```python
no_identity_theft = df[~(df['Crm Cd Desc'] == 'THEFT OF IDENTITY')]
```

```python
new_hourly_crime = no_identity_theft['TIME OCC'].dt.hour.value_counts().sort_index().reset_index()
new_hourly_crime
```

```python
plt.bar(new_hourly_crime['TIME OCC'], new_hourly_crime['count'])
plt.title('Crime Count by Time of Day')
plt.show()
```

Upon further analysis, it seems that the default value to note time of crime is at 12:00. That could be a reason why 12:00 has a high count. 

```python
no_identity_theft[no_identity_theft['TIME OCC'].dt.hour == 12]['Crm Cd Desc'].value_counts()
```

For example, stolen vehicle, there might not be an accurate time to know when a vehicle is stolen.
